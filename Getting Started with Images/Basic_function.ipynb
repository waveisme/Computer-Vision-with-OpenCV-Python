**Reading Images using OpenCV**
* OpenCV can read different types of images (JPG, PNG, ...), grayscale images, color images, images with alphal channel
* Function syntax: cv2.imread(filename, flags)
** filename:a path
** flags: type of image, default 1 - color images, 0 - grayscale images, -1 - alphal channel

```python
# read image as gray scale
img = cv2.imread("checkerboard_18x18.png",0)
# pixel values - element of a 2D numpy array.
# Each pixel has 8 bits [0,255]
print(img)
```

**Display Image attributes**

```python
#print the size of image
print("Image size is",img.shape)
#Data-type of image
print("Data type of image is",img.dtype)
```

**Display images using matplotlib**

**Working with color images**
A (digital) color image is a digital image that includes color information for each pixel. ... A color image has three values (or channels) per pixel and they measure the intensity and chrominance of light.

```python
#Read and display Coca-Cola logo directly by Image of IPython
Image(filename = "coca-cola-logo.png")
```

```python
img3 = cv2.imread("coca-cola-logo.png",1)
print(img3)
```

```python
print("Image size is",img3.shape)
print("Data type of image is",img3.dtype)
```

```python
#using matplotlib
plt.imshow(img3)
```

This is because matplotlib expects the image in RGB format, OpenCV stores images in BGR format. Reverse the channels of the image

```python
img3_reversed = img3[:,:,::-1]
plt.imshow(img3_reversed)
```

**Splitting and Merging Color Channels**
* cv2.split(): Divides a multi-channel array into single-channel
* cv2.merge(): make several arrays to make a singel multi-channel array

```python
#split the image into the B,G,R components
img_bgr = cv2.imread("New_Zealand_Lake.jpg",1)
b,g,r =cv2.split(img_bgr)

#show the channel
plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(r); plt.title("Red Channel")
plt.subplot(142); plt.imshow(g); plt.title("Green Channel")
plt.subplot(143); plt.imshow(b); plt.title("Blue Channel")

#merge the serveral channel into a BGR image
img_merge = cv2.merge((b,g,r))
# show
plt.subplot(144); plt.imshow(img_merge[:,:,::-1]); plt.title("Merged Output")
```

**Converting to different color spaces**
