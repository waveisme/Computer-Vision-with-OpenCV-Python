* Draw lines
* Draw circles
* Draw rectangles
* Add text

**Import Libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import matplotlib
from IPython.display import Image
matplotlib.rcParams['figure.figsize'] = (9.0,9.0)
```

```python
#Read an image
img = cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)
#Display
plt.imshow(img[:,:,::-1])
```

**Drawing a line**
* cv2.line
* img = cv2.line(img,pt1,pt2, color,thickness, lineType,shift)
** img: output image
** img: Image on which we will draw a line
** pt1: First point of the line segment
** plt2: Second point of the line segment
** color: color of the line which will be drawn (B,G, R)
** thickness:Integet specifying the line thickness. default value is 1
** lineType: Type of the line. Default value is 8.
** cv2.LINE_AA (antialiased or smooth line)

```python
img_line = img.copy()
cv2.line(img,(200,400),(400,400),(255,200,100), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(img[:,:,::-1])
```

**Drawing a Circle**
 img = cv2.circle(img,center,radius,color,thicknesss,lineType,shift)
 * img: Image on which we will draw a line
 * center: center of the circle
 * radiuse: radius of the circle
 * color: color of the circle which will be drawn

```python
img2 = cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)
cv2.circle(img2,(200,200),150,(0,0,255),10,lineType = cv2.LINE_AA)
plt.imshow(img2[:,:,::-1])
```

**Drawing a rectangle**
 use cv2.rectangle
*img = cv2.rectangle(img,pt1,pt2,color,thickness,lineType,shift)
* pt1: top_left vertex
* pt2: bottom - right

```python
img = cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)
cv2.rectangle(img,(500,100),(700,600),(255,0,255),thickness = 5,lineType = cv2.LINE_AA)
plt.imshow(img[:,:,::-1])
```

**Adding Text**
img = cv2.putText(img,text,org,fontFace,fontScale,color,thickness,lineType,bottomLeftOrigin)
* text: text string to be written
* org: bottom-left corner of the text string in the image
* fontFace: Font type
* fontScale: size

```python
img = cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)
text = "Apollo 11 Saturn V Launch, July 16, 1969"
cv2.putText(img,text,(200,700),cv2.FONT_HERSHEY_PLAIN,2.3,(0,255,0),2,cv2.LINE_AA)
plt.imshow(img[:,:,::-1])
```
