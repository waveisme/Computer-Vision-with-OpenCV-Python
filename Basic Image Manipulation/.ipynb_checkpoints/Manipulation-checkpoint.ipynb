How to Perform image transformations including
* Accessing and manupulating images pixels
* Image resizing
* Cropping
* Flipping

**Import libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
from IPython.display import Image
```

**Read image as a gray scale**

```python
img = cv2.imread("checkerboard_18x18.png",0)
print(img)
plt.imshow(img,cmap = 'gray')
```

**Accessing Individual Pixels**
* Use matrix notation such as matrix[r,c] - row, column (0 - index)

```python
#First pixel of the first black box
print(img[0,0])
#First white pixel to the right of the first black
print(img[0,6])
```

**Modifying Image Pixels**

```python
#copy image: copy()
img_copy = img.copy()
img_copy[2,2] = 200
img_copy[2,3] = 200
img_copy[3,2] = 200
img_copy[3,3] = 200
plt.imshow(img_copy,cmap = 'gray')
print(img_copy)
```

```python
img_copy[2:5,2:5] = 200
plt.imshow(img_copy,cmap = 'gray')
```

**Cropping Images**
 select a specific (pixel) region of the image

```python
img_bgr = cv2.imread("New_Zealand_Boat.jpg",cv2.IMREAD_COLOR)
img_rgb = img_bgr[:,:,::-1]
plt.imshow(img_rgb)
```

```python
#crop out the middle region of the image
img_crop = img_rgb[200:400,300:600]
plt.imshow(img_crop)
```

**Resizing Images**
* resize: dst = resize(src, dsize[,fx[,fy[,interpolatint]]]])
** dst: output image
** src input image
** dsize output image size
Optional arguments that are often used include:

    fx: Scale factor along the horizontal axis; when it equals 0, it is computed as (𝚍𝚘𝚞𝚋𝚕𝚎)𝚍𝚜𝚒𝚣𝚎.𝚠𝚒𝚍𝚝𝚑/𝚜𝚛𝚌.𝚌𝚘𝚕𝚜
    fy: Scale factor along the vertical axis; when it equals 0, it is computed as (𝚍𝚘𝚞𝚋𝚕𝚎)𝚍𝚜𝚒𝚣𝚎.𝚑𝚎𝚒𝚐𝚑𝚝/𝚜𝚛𝚌.𝚛𝚘𝚠𝚜


```python
# Specifying Scaling factor using fx and fy
resize_img = cv2.resize(img_crop,None,fx = 2, fy = 2)
plt.imshow(resize_img)
```

```python
#Specifying exact size of the output image
resize_img2 = cv2.resize(img_crop,dsize = (100,200), interpolation = cv2.INTER_AREA)
plt.imshow(resize_img2)
```

```python
#Resize whiel maitaining aspect ratio
 #Method 2: Using 'dsize'
desired_width = 100
aspect_ratio = desired_width / img_crop.shape[1]
desired_height = int(img_crop.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

# Resize image
resized_cropped_region = cv2.resize(img_crop, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
resize_cropped_region2 = resized_cropped_region[:,:,::-1]
cv2.imwrite("1.png",resize_cropped_region2)
Image(filename = "1.png")
```

**Fipping Images**
* Filp the array in one of three different ways (row and column indices are 0 - based)
* dst = cv.flip(scr,flipCode)
* 0 flipping around the x-axis
* 1 flipping around the y-axis
* -1 flipping around both axes

```python
img_bgr = cv2.imread("New_Zealand_Boat.jpg",cv2.IMREAD_COLOR)
img_rgb = img_bgr[:,:,::-1]
img1 = cv2.flip(img_rgb,0)
img2 = cv2.flip(img_rgb,1)
img3 = cv2.flip(img_rgb,-1)
#display
plt.figure(figsize = (20,5))
plt.subplot(141);plt.imshow(img1); plt.title("X-axis")
plt.subplot(142);plt.imshow(img2); plt.title("Y-axis")
plt.subplot(143);plt.imshow(img3); plt.title("X-Y")
plt.subplot(144);plt.imshow(img_rgb); plt.title("Original")
```
